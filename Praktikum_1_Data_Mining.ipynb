```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# inisialisasi data
categories = ['A', 'B', 'C', 'D']
values = [5, 12, 8, 15]
```

```python
# Bar Chart
# setting data ke dalam template
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title("Bar Chart Example")

# menampilkan gambar
plt.show()
```

```python
# Line Chart
# Inisialisasi data
x_values = [1,2,3,4,5]
y_values = [10,12,5,8,15]

# setting data ke template
plt.plot(x_values, y_values)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Line Chart Example")

plt.show()
```

```python
# Scatter Plot
# Inisialisasi data
x_values = [1,2,3,4,5]
y_values = [10,12,5,8,15]

# setting data ke template
plt.scatter(x_values, y_values)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Scatter Plot Example")

plt.show()
```

```python
# Pie Chart
# Inisialisasi data
labels = ['Category A', 'Category B', 'Category C']
sizes = [25, 40, 35]

# Setting data ke dalam template
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart Example")

plt.show()
```

```python
# Box Plot
# Load dataset
data = sns.load_dataset('tips')

#setting data ke template
sns.boxplot(data=data, x='day', y='total_bill')
plt.title("Box Plot Example")

plt.show()
```

```python
# Heatmap
# Load dataset
data = sns.load_dataset('flights').pivot_table(index='month', columns='year', values='passengers')

#setting data ke dalam template
plt.figure(figsize=(12,5))
sns.heatmap(data, annot=True, cmap='coolwarm')

plt.show()
```

```python
# Diagram Violin
# Load dataset
data = sns.load_dataset('iris')

# Setting data ke dalam template
sns.violinplot(data, x='species', y='sepal_length')
plt.title("Violin Plot Example")

plt.show()
```

```python
# Pandas Merge
# Inisialisasi data
df1 = pd.DataFrame({'ID':[1,2,3], 'Name':['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID':[1,2,4], 'Age':[25,30,22]})

# Setting ke dalam template
merge_df = pd.merge(df1, df2, how='inner')

# Menampilkan Hasil
merge_df
```

```python
# Join
# Inisialisasi data
df1 = pd.DataFrame({'ID':[1,2,3], 'Name':['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'Age':[25,30,22]}, index=[1,2,4])

# Setting data ke dalam template
joined_df = df1.join(df2, on='ID', how='left')

# Menampilkan hasil integrasi
joined_df
```

```python
# Concatenation (Concat)
# Inisialisasi data
df1 = pd.DataFrame({'ID':[1,2,3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID':[4,5,6], 'Name': ['David', 'Eva', 'Frank']})

# Setting data ke dalam template
concatenated_df = pd.concat([df1, df2], axis=0)

concatenated_df
```

```python
# Pembersihan Data Numerik
data = sns.load_dataset('tips')

# Menampilkan isi dataset
data.head()
```

```python
# Mengidentifikasi Nilai Hilang (Missing Value)

missing_values = data.isnull().sum()
print("Jumlah data hilang: ")
print(missing_values)
```

```python
# Menangani data yang hilang dengan nilai mean
data.fillna(data.mean(), inplace=True)
```

```python
# Menangani duplikasi data
data.drop_duplicates(inplace=True)
```

```python
# Mengidentifikasi outlier dengan boxplot
sns.boxplot(data, y='total_bill')
```

```python
# Menangani outlier dengan menghapusnya menggunakan z-score
data['total_bill_Z'] = (data['total_bill']-data['total_bill'].mean())/data['total_bill'].std()
data.head()
```

```python
# !pip install sqlalchemy

data_no = data[(data.total_bill_Z<2)]
data_no.head()
```

```python
# Menampilkan hasil data setelah hapus outlier
sns.boxplot(data_no, y='total_bill')
```

```python
# Normalisasi/Standardisasi

from sklearn.preprocessing import MinMaxScaler
data = sns.load_dataset('iris')

# misal : menggunakan minmaxscaler

scaler = MinMaxScaler()
data.head()
```

```python
# Menampilkan hasil normalisasi
data.drop('species', axis=1, inplace=True)
data_sc = scaler.fit_transform(data)
data = pd.DataFrame(data_sc, columns=data.columns)

data.head()
```
